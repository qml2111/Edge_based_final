```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_path ='/home/tanker/Downloads/canny_edge_detector-master/faces_imgs/6.jpg'
img = cv2.imread(img_path)
```

```python
plt.imshow(img[:,:,::-1])
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_blur = cv2.GaussianBlur(gray_image, (5,5),0)
plt.imshow(gray_image, cmap='gray', vmin = 0, vmax = 255)
```

```python

Kx = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
Ky = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
```

```python
sobel_GX = cv2.filter2D(src = img_blur, ddepth=-1, kernel=Kx)
sobel_GY = cv2.filter2D(src = img_blur, ddepth=-1, kernel=Ky)
```

```python
sobel_XY = cv2.addWeighted(sobel_GX, 0.5, sobel_GY, 0.5,0)
```

```python
plt.imshow(sobel_XY,cmap='gray')
```

```python
def mse(X, Y):
    xh, xw = X.shape
    yh, yw = Y.shape
    
  
    assert ((xw == yw) or (xh == yh)), \
    "Arguments dimensions are not equals: %s not equal to %s" % (str(xw), str(yw))
    a = [((X[i,j] - Y[i,j])**2) for i in range(xh) for j in range(xw)]
    return np.sum(a) / (xw * xh)
```

```python
x = mse(gray_image,sobel_XY)
```
